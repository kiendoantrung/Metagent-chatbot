```python
import requests
from bs4 import BeautifulSoup
import json

base_url = 'https://metagent.vn/lookbook/'

lookbook_ids = [
    'the-first-step-1',
    'touching-winter-2',
    'off-the-grid-3',
    'the-seasons-symphony-4',
    'icy-summer-5',
    'simple-pleasure-collection-7',
    'floating-light-8'
]

data = []

for id in lookbook_ids:
    url = base_url + id
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    lookbooks = soup.find_all('div', class_='lookbook-content')

    for lookbook in lookbooks:
        # Extract lookbook title and description
        lookbook_title = lookbook.find('h3').text.strip()
        lookbook_desc = lookbook.find('p').text.strip()

        products = soup.find_all('div', class_='product')

        lookbook_data = {
            'lookbook': lookbook_title,
            'infor_lookbook': lookbook_desc,
            'products': []
        }

        for product in products:
            title = product.find('h3', class_='title-product').text.strip()
            price_element = product.find('div', class_='price-product')
            price = price_element.text.strip().split('\n')[-1] if price_element else ''

            thumb = product.find('div', class_='thumb-product')
            img_url = thumb.find('img')['src']
            product_url = thumb.find('a')['href']

            # Make detail page request
            detail_response = requests.get(product_url)
            detail_soup = BeautifulSoup(detail_response.text, 'html.parser')

            # Get paragraph texts
            tab_content_div = detail_soup.find('div', class_='tab-content')
            def filter_paragraphs(tag):
                return tag.name == 'p' and not tag.find('strong')

            filtered_p_tags = tab_content_div.find_all(filter_paragraphs)
            infor = ' '.join([p.text.strip() for p in filtered_p_tags])

            lookbook_data['products'].append({
                'url': product_url,
                'title': title,
                'price': price,
                'img_url': img_url,
                'infor': infor
            })

        data.append(lookbook_data)

# Print the data as JSON
print(json.dumps(data, indent=2, ensure_ascii=False))
```
